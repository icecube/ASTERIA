# Asymmetric Uncertainties

A model for asymmetric uncertainties [(Eq 21 in Barlow, 2003)](https://www.slac.stanford.edu/econf/C030908/papers/WEMT002.pdf)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1, 1, figsize = (3,5))
m = 5.0 # central value, an example
a = np.linspace(0, 20, 100)

# Negative and Positive uncertainties
sig_neg = 2.58
sig_pos = 1.92

gamma = (sig_pos*sig_neg)/(sig_pos-sig_neg)
beta = sig_pos/sig_neg
lnL = (-0.5*np.power((np.log(1+(m-a)/gamma)/np.log(beta)), 2))
#I have no idea why I used (m-a), but it worked

ax.plot(a, lnL)
ax.set(xlabel = "m", ylabel = "lnL")
fig.tight_layout()
```
